```python
Sex=pd.get_dummies(t_data['sex'],drop_first=True)
```

```python
Embark=pd.get_dummies(t_data['embarked'],drop_first=True)
```

```python
Pclass=pd.get_dummies(t_data['pclass'])
```

```python
t_data=pd.concat([t_data,Sex,Embark,Pclass],axis=1)
```

```python
t_data.head()
```

```python
t_data.drop(['pclass','sex','embarked','ticket','body','boat','name','home.dest'],axis=1,inplace=True)
```

```python
t_data.head()
```

```python
t_data.drop(['cabin'],axis=1,inplace=True)
```

```python
t_data.head()
```

```python
t_data.describe()
```

```python
t_data=t_data.fillna(t_data.median())
t_data.head()
```

# Testing of data

```python
X=t_data.drop(['survived'],axis=1)
y=t_data['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression  
```

```python
logisticmodel=LogisticRegression()
```

```python
t_data.head()
```

```python
t_data.describe()
```

```python
logisticmodel.fit(X_train,y_train)
t_data.head()
```

```python
prediction=logisticmodel.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
classification_report(y_test,prediction)
```

```python
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```
